```
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

music = pd.read_csv('top10s.csv', engine='python', index_col=0)
music.rename(columns={"pop":"popularity"}, inplace=True)
plt.style.use('ggplot')

music.head(20)

```

```
music.loc[:,['title','artist','dnce','nrgy','popularity']].sort_values("popularity",ascending=False).head(20)
```

```
music.loc[:,['title','artist','dnce','nrgy','popularity']].sort_values("dnce",ascending=False).head(20)
```

```
music.artist.describe()
```

```
music.popularity.describe()
```

```
music.popularity.value_counts().sort_index().plot.bar(edgecolor='black')
```

```
music[music.popularity > 55].popularity.plot.hist(edgecolor="black")
```

```
music[music.nrgy > 50].plot.hexbin(x='nrgy',y='dnce',gridsize=12)
```

```
music[music.nrgy > 40].plot.scatter(x='nrgy',y='dnce')
```
